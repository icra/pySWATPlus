```python
from pySWATPlus import TxtinoutReader
import re
```

```python
txtinout_folder = '/mnt/c/users/joans/onedrive/Escriptori/icra/muga_windows'    #Here you have to put the path to your  txtinout folder, with the SWAT.exe 
filename =  'plants.plt'
```

```python
#The folder pointed by txtinout_folder must have a .exe file to run the simulation
reader = TxtinoutReader(txtinout_folder)
```

```python
#Read plants.plt file
plants_reader = reader.register_file(filename, 
                    has_units=False,
                    index='name')
```

```python
plants_reader.header_file
```

```python
plants_reader.df.head()
```

```python
reader.set_beginning_and_end_year(2009, 2010)
```

```python
reader.set_warmup(1)
```

```python
#Enable the 'channel_sd' in the print.prt file (the other ones are mantained as they are)
reader.enable_object_in_print_prt(obj='channel_sd', daily=True, monthly=False, yearly=False, avann=False)
```

```python
%%capture captured_output

#Run the model with the default settings. With show_output=True, the output of the model will be printed in the console.
#We capture it using the magic command %%capture to avoid printing it.
txtinout_path = reader.run_swat(show_output=True)
```

```python
#If can print now the first 15 lines of the output, to see that the model has run correctly
print("\n".join(captured_output.stdout.splitlines()[:15]))
```

```python
#and txtinout path has now the path to the folder where the simulation has been run
txtinout_path
```

```python
#For reading the results of the simulation, we can use the same reader object because the output files are in the same folder. Otherwise, we would need to create a new reader object with the path to the output folder.
channel_sd = reader.register_file('channel_sd_day.txt', 
                    has_units=True)
channel_sd.df.head()
```

```python
#For modifying the input files, either modify the stored dataframe and save it, and then run the model again...
plants_reader.df.loc['agrc', 'bm_e'] = 40
plants_reader.df.head()
plants_reader.overwrite_file()
```

```python
#Or directly the model specifying the modifications
reader.run_swat(params = {'plants.plt': ('name', [('agrc', 'bm_e', 50)])}, show_output=False)
```

```python
#Or for multiple modifications, you can use a regular expression. In this case, all the land cover/plants starting with 'a' will have the 'bm_e' value set to 50
txtinout_path = reader.run_swat(params={'plants.plt': ('name', [(re.compile('^a', re.IGNORECASE), 'bm_e', 50)])}, show_output=False)
```

```python
#You can also run multiple simulations on parallel. Notice that in this case, each simulation will be run in a different folder, so the output files will be stored in different folders.
txtinout_paths = reader.run_parallel_swat(params = [{'plants.plt': ('name', [('bana', 'bm_e', 45)])}, {'plants.plt': ('name', [('bana', 'bm_e', 40)])}], n_workers = 2, parallelization='threads')
txtinout_paths
```
